```python
#5. Realiza los cálculos necesarios para conocer la edad promedio de los participantes
#y si hay más hombres o mujeres contagiados.

import os

def calculate_statistics(filename):
    total_edad = 0
    Cuenta_Masculino = 0
    Cuenta_Femenino = 0

    with open(filename, 'r') as file_handler:
        next(file_handler)

        for line in file_handler:
            values = line.strip().split(',')
            age = int(values[3])
            sex = values[2]
            total_edad += age

            if sex == 'MASCULINO':
                Cuenta_Masculino += 1
            elif sex == 'FEMENINO':
                Cuenta_Femenino += 1

    edad_promedio = total_edad / (Cuenta_Masculino + Cuenta_Femenino)

    return edad_promedio, Cuenta_Masculino, Cuenta_Femenino


filename = '/content/casos_confirmados.csv'
edad_promedio, Cuenta_Masculino, Cuenta_Femenino = calculate_statistics(filename)


# 6.Imprime la respuesta a las siguientes preguntas:
 #¿Cuál es la edad promedio de los pacientes?
# ¿Hay más hombres o mujeres contagiados?


print(f"¿Cual es la Edad Promedio de los Pacientes en Mexico? \nEn el estudio se identifico que la Edad promedio de pacientes en Mexico fue de : {edad_promedio}")
print('')
print('')
print('¿Hay más hombres o mujeres contagiados? \nEl estudio demostro que fueron mas los Hombres Contagiados tal y como muestran las siguietes Cifras')
print('')
print('')
print(f"Cantidad de Hombres contagiados: {Cuenta_Masculino}")
print(f"Cantidad de Mujeres contagiadas: {Cuenta_Femenino}")
```
