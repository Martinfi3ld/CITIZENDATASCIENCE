```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#a. extraer informacion
vaccines = pd.read_csv('/contryvaccine.csv', parse_dates=['date'])
vaccines.head()
```

```python
#b. Mostrar la estructura y tipos de datos de cada columna para identificar qué

#operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

vaccines.info()
```

```python

#c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines,

#en otras palabras, agrupe por vaccines y realice la sumatoria).

Suma_por_compañia = vaccines[['vaccines', 'daily_vaccinations']].groupby('vaccines').sum()
Suma_por_compañia
```

```python
#d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

print('Cantidad de vacunas aplicadas en todo el mundo: \n', vaccines['daily_vaccinations'].sum())
```

```python
#e. Calcular el promedio de vacunas aplicadas por país.

mean_by_contry = vaccines[['country', 'daily_vaccinations']].groupby('country').mean()
mean_by_contry
```

```python
#f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

print('vacunas aplicadas el 2021-91-29: ', vaccines[vaccines['date'] == '2021-01-29']['daily_vaccinations'].sum())
```

```python
#g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias)
#con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

ConDiferencias = vaccines.copy()
ConDiferencias['diferencias'] = vaccines['daily_vaccinations'] - vaccines['daily_vaccinations_raw']
ConDiferencias

```

```python
vaccines.head(0) #ESTE ES UN VALIDADOR DE QUE SOLO SE EJECUTARA LA COLUMNA DE DIFERENCIAS EN EL DATAFRAME conDiferencias y no en el de vaccines
```

```python
#h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

print('Periodo de tiempo entre primer registro y ultimo: \n ', vaccines['date'].max() - vaccines['date'].min())
```

```python
#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada
# (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , )

ConCantidad = vaccines.copy()
ConCantidad['canVac'] = vaccines['vaccines'].str.split(',').str.len()
ConCantidad[['country','canVac']] #LO MUESTRO CON DOS COLUMNAS PARA QUE SEA MAS FACIL LEER EL RESULTADO
```

```python
#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

antes20 = vaccines[vaccines['date'] < '2020-12-20']
antes20['date'].unique() #UTILIZO LA FUNCION .UNIQUE PARA DEMOSTRAR QUE EL NUEVO DATA FRAME TIENE LAS FECHAS MENOR A LO INDICADO EN EL EJERCICIO
```

```python
#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

pfizer = vaccines[vaccines['vaccines'].str.contains('Pfizer')]
pfizer['aplico Pfizer'] = pfizer['vaccines'].str.contains('Pfizer') #SE APLICO PARA DEMOSTRAR QUE SI SE PLICO Pfizer
pfizer[['country','vaccines','aplico Pfizer']]
```

```python
#l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel
#denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    ConDiferencias.to_excel(writer, sheet_name='conDiferencias')
    ConCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')


```
